```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv('/content/test.csv')
```

```python
# Display the first few rows to get an overview of the data structure.

titanic_df.head(20)
```

```python
# Check the dimensions of the dataset (number of rows and columns).

titanic_df.shape
```

```python
# Identify missing values and decide on strategies for handling them.

titanic_df.isna().sum() #how many empty values are there

titanic_df.notna().sum()  #how many not-empty values are there

titanic_df.notna().all()  #you can check if all data has been filled in/if smt is missing somewhere

titanic_df.notna().sum()  #how many not-empty values are there
```

```python
# Compute basic statistics (e.g., mean, median, min, max) for numerical columns.

titanic_df['Age'].mean()
titanic_df['Age'].median()
titanic_df['Age'].min()
titanic_df['Age'].max()
```

```python
# Analyze categorical variables (e.g., value counts).

titanic_df['Sex'].value_counts()
```

```python
# Choose one column of the dataset that interests you
# Use Matplotlib pyplot to create a meaningful visualization:
# Select an appropriate plot type (e.g., bar chart, histogram, pie chart).
# Label axes, add a title. Ensure the visualization is clear, informative, and visually appealing.

plt.hist(titanic_df['Age'], bins= 12, color='lightgreen', edgecolor='black', alpha=0.5)
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.title('Ages of travellers on Titanic')

plt.axvline(titanic_df['Age'].mean(), linestyle='dashdot', color='blue')
plt.axvline(titanic_df['Age'].min(), linestyle='dashed', color='red')
plt.axvline(titanic_df['Age'].max(), linestyle='dashed', color='orange')

plt.grid(linestyle='--', alpha=0.7)
```
